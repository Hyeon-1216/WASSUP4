## QUIZ: 네이버 증권 주요뉴스의 제목과 url을 모두 추출하기

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
PATH = "https://finance.naver.com/"
resp = requests.get(PATH)
html_src = resp.text
soup = BeautifulSoup(html_src, 'lxml')
```

```python
taglist = soup.select('.section_strategy ul a')
taglist
```

```python
title_list = []
url_list = []
for i in taglist:
    title =  i.text
    url = i['href']
    title_list.append(title)
    url_list.append(PATH[:-1] + url)
zipped = zip(title_list,url_list)

import pandas as pd
df = pd.DataFrame(zipped, columns = ['title', 'url'])
df
```

# 멜론차트, 노래제목고 가수 수집하기

https://www.melon.com/chart/index.htm

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
PATH = 'https://www.melon.com/chart/index.htm'
resp = requests.get(PATH)
resp
```

```python
requests.get?
```

```python
info = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36'}
info
```

```python
resp = requests.get(PATH, headers = info)
resp
```

```python
html_src = resp.text
html_src
soup = BeautifulSoup(html_src, 'lxml')
```

```python
title_list = []
singer_list = []

title = soup.select('.ellipsis.rank01 a')
singer = soup.select('ellipsis.rank02 a')

for i in title:
    title_list.append(i.text)
for j in singer:
    singer_list.append(j.text)
    
singer_list
```

```python
title= [i.text for i in soup.select('.ellipsis.rank01 a')]
title
```

```python
singer= [i.text for i in soup.select('.checkEllipsis')]
singer
```

```python
rank = list(range(1,101))
rank
```

```python
zipped = zip(rank, title, singer)
df = pd.DataFrame(zipped, columns = ['순위', '노래', '가수'])
df
df.set_index('순위')

df.to_csv('output/melon.csv')
```
